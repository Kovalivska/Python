```python
from faker import Faker



# Create a Faker instance

fake = Faker()



# Generate some fake data

print("Name:", fake.name())

print("Address:", fake.address())

print("Email:", fake.email())

print("Phone:", fake.phone_number())

print("Job:", fake.job())



# Generate a list of fake names

print("\nFake Names:")

for _ in range(5):  # Generate 5 names

    print(fake.name())
```

```python
from faker import Faker
import pandas as pd

# Create a Faker instance
fake = Faker()

# Number of rows to generate
num_rows = 100

# Create an empty list to hold the rows
data = []

# Generate fake data using a for loop
for _ in range(num_rows):
    row = {
        "Name": fake.name(),
        "Address": fake.address().replace('\n', ', '),  # Replace newlines in addresses
        "Email": fake.email(),
        "Phone": fake.phone_number(),
        "Job": fake.job()
    }
    data.append(row)

# Create a DataFrame
df = pd.DataFrame(data)

# Display the DataFrame
df.head()
```
