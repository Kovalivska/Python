```python
# Step 1: Importing Libraries and Dataset
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

# Load Iris dataset
iris = load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)

# Step 2: Explore and Prepare the Data
print(df.head())

# Step 3: Scaling the Data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(df)

# Step 4: Applying K-Means Clustering
kmeans = KMeans(n_clusters=3)

# Step 5: Training the K-Means Model
kmeans.fit(X_scaled)

# Step 6: Assigning Cluster Labels
df['Cluster'] = kmeans.labels_

# Step 7: Visualizing the Clusters
plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c=df['Cluster'], cmap='viridis')
plt.xlabel('Sepal Length (scaled)')
plt.ylabel('Sepal Width (scaled)')
plt.title('K-Means Clustering of Iris Data')
plt.show()

# Step 8: Evaluating the Model with Silhouette Score
score = silhouette_score(X_scaled, df['Cluster'])
print(f'Silhouette Score: {score}')
```

To evaluate the quality of the clusters formed by DBSCAN, we’ll use the Silhouette Score again. As a reminder, this metric measures how similar points in a cluster are to each other compared to points in other clusters. It ranges from -1 to 1, where:
A score close to 1 means points are well-clustered.
A score close to 1 means points might be in the wrong cluster.
Here’s how to calculate the silhouette score, excluding noise points:help(pd.DataFrame)

What Does This Mean?
DBSCAN's Behavior
DBSCAN (Density-Based Spatial Clustering of Applications with Noise) clusters data based on:
Density: A minimum number of points (min_samples) required to form a cluster within a neighborhood of a given radius (eps).
Neighborhood: Defined by the distance between points (eps).
In your case, none of the data points meet the density requirements to form a cluster. As a result, all points are classified as noise (-1).
Possible Reasons for Noise
eps Too Small: The neighborhood radius might be too small to include sufficient points to form a cluster.
min_samples Too High: The minimum number of points required to form a cluster might be too high, causing DBSCAN to treat small groups of points as noise.
Scaled Data Characteristics: The scaling of your data may not be appropriate for the distance metric used by DBSCAN (default is Euclidean).

How to Fix This?
Tune Hyperparameters
Increase the value of eps (neighborhood radius).
Decrease the value of min_samples (minimum points for a cluster).

Challenge 5: Train DBSCAN Models with Different Parameters
Do you remember that we said that DBSCAN’s performance heavily depends on parameters like eps and min_samples? You will experiment with different values for eps and min_samples to train multiple DBSCAN models and evaluate each one.
In the precode ranges for eps and min_samples are provided, along with code to visualize the clusters. Your goal is to:
Train DBSCAN models using different combinations of eps and min_samples.
Exclude noise
Calculate the silhouette score for each model


```python
# Set ranges for eps and min_samples
eps_values = [2.169, 3, 3]
min_samples_values = [14, 39, 26]


# Import silhouette_score
from sklearn.metrics import silhouette_score

# Iterate over combinations of eps and min_samples
for eps in eps_values:
    for min_samples in min_samples_values:
        # Initialize DBSCAN with current parameters
        dbscan = DBSCAN(eps=eps, min_samples=min_samples)

        # Fit the model
        dbscan.fit(scaled_data)

        # Get cluster labels
        labels = dbscan.labels_

        # Exclude noise points (-1 labels)
        filtered_data = scaled_data[labels != -1]
        filtered_labels = labels[labels != -1]

        # Calculate silhouette score
        if len(set(filtered_labels)) > 1:  # Ensure more than one cluster
            silhouette_avg = silhouette_score(filtered_data, filtered_labels)
            print(f"eps: {eps}, min_samples: {min_samples}, Silhouette Score: {silhouette_avg}")
        else:
            print(f"eps: {eps}, min_samples: {min_samples}, Not enough clusters for silhouette score")
```
