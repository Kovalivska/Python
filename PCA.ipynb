```python

import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.datasets import load_iris

# Load the Iris dataset
iris = load_iris()
X = iris.data  # Features
y = iris.target  # Labels

from sklearn.preprocessing import StandardScaler

# Standardize the data (zero mean, unit variance)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Initialize PCA to reduce the dataset to 2 components
pca = PCA(n_components=2)

# Fit PCA to the scaled data and transform it
X_pca = pca.fit_transform(X_scaled)

# Plot the two principal components
plt.figure(figsize=(8, 6))
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y, cmap='plasma', edgecolor='k', s=100)
plt.title('PCA on Iris Dataset')
plt.xlabel('First Principal Component')
plt.ylabel('Second Principal Component')
plt.show()

# Print the explained variance ratio of each component
print(pca.explained_variance_ratio_)

# Print the cumulative variance explained by the two components
print(f"Cumulative explained variance: {np.sum(pca.explained_variance_ratio_):.2f}")

# Initialize PCA to keep all components
pca_full = PCA()

# Fit PCA to the scaled data
pca_full.fit(X_scaled)

# Plot the cumulative explained variance
plt.figure(figsize=(8, 6))
plt.plot(np.cumsum(pca_full.explained_variance_ratio_), marker='o', linestyle='--')
plt.title('Explained Variance by PCA Components')

```

#PCA: Practice, Part 1

##1: Standardizing the Dataset


You have been provided with the Wine dataset, and your task is to standardize the data before applying PCA. Standardization ensures that all features have the same scale, which is crucial for PCA.
Instructions:
Import and use the StandardScaler to standardize the data.
Fit the scaler to the data and transform it.


```python
from sklearn.preprocessing import StandardScaler # Import Standard Scaler
from sklearn.datasets import load_wine

# Load the Wine dataset
wine = load_wine()
X = wine.data

# Initialize the scaler and standardize the dataset
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

##2. Apply PCA


Now, we can apply PCA. Letâ€™s reduce the dataset from 4 dimensions down to 2, so we can easily visualize the results.
Here, we specified n_components=3 to reduce the dataset to 3 dimensions.

```python
from sklearn.decomposition import PCA
# Initialize PCA to reduce the dataset to 3 components
pca = PCA(n_components=3)

# Fit PCA to the scaled data and transform it
X_pca = pca.fit_transform(X_scaled)
```

##3: Visualizing the PCA Components


After applying PCA, itâ€™s time to visualize the two principal components of the Wine dataset.
Instructions:
Create a scatter plot of the two principal components using matplotlib. Your goal is to see the first and the second principal components by their indices.

```python
from mpl_toolkits.mplot3d import Axes3D

# Create a 3D scatter plot
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
scatter = ax.scatter(
    X_pca[:, 0], X_pca[:, 1], X_pca[:, 2],
    c=y, cmap='plasma', edgecolor='k', s=100
)

# Add labels and title
ax.set_title('PCA on Wine Dataset (3D)')
ax.set_xlabel('First Principal Component')
ax.set_ylabel('Second Principal Component')
ax.set_zlabel('Third Principal Component')

# Add color bar
cbar = plt.colorbar(scatter, ax=ax, pad=0.1)
cbar.set_label('Wine Type')

plt.show()
```

##4: Checking Explained Variance


After applying PCA, you need to check how much variance each principal component captures from the original data.
Instructions:
Access the explained variance for each component and print it
Calculate and print the cumulative variance explained by the two components.


```python
import numpy as np
# Print the explained variance ratio of each component
explained_variance = pca.explained_variance_ratio_  # Access the explained variance ratio
print(explained_variance)

# Calculate and print cumulative explained variance
cumulative_variance = np.sum(explained_variance[:3] )  # Sum the explained variance
print(f"Cumulative explained variance: {cumulative_variance:.2f}")
```

## 5. Plot the cumulative explained variance

```python
# Initialize PCA to keep all components
pca_full = PCA()

# Fit PCA to the scaled data
pca_full.fit(X_scaled)

# Plot the cumulative explained variance
plt.figure(figsize=(8, 6))
plt.plot(np.cumsum(pca_full.explained_variance_ratio_), marker='o', linestyle='--')
plt.title('Explained Variance by PCA Components')
```
